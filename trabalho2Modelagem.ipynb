# Enunciado

## Projeto: Validação Empírica da Lei de Zipf — Livros escritos por humanos (CN) vs. Livros escritos por Inteligência Artificial (CA)

### 1. Visão Geral

O objetivo deste trabalho é realizar um experimento computacional para verificar se textos gerados por Grandes Modelos de Linguagem (LLMs) seguem a Lei de Zipf da mesma forma que a literatura humana clássica.

A **Lei de Zipf** estabelece que a frequência de uma palavra em um corpus é inversamente proporcional ao seu ranking:

$$f(r) \approx \frac{C}{r^{\alpha}}$$

Os alunos deverão desenvolver um pipeline que:

1.  Coleta dados reais (Corpus Natural - CN)
2.  Gera dados sintéticos (Corpus Artificial - CA) via API
3.  Realiza uma análise estatística comparativa dos coeficientes de decaimento ($\alpha$)

### 3. Fases do Trabalho (Roteiro de Implementação)

#### Fase 1: Coleta e Limpeza (CN)

*   **Fonte**: Download automatizado de obras em prosa do **Project Gutenberg**, utilizando o formato de URL `https://gutenberg.org/cache/epub/<ID>/pg<ID>.txt`. Uma lista estável de `GUTENBERG_IDS` será utilizada para seleção. (FEITO)
*   **Processamento**: Limpar o texto removendo pontuações e números.
*   **Normalização**: Extrair exatamente `N` palavras (ex: 25.000) de cada livro para garantir uniformidade. Livros muito curtos ou com falhas de download serão descartados.
*   **Alvo**: Coletar no minímo **100** livros, todos em um mesmo idioma.

#### Fase 2: Geração de Corpus via API (CA)

*   **Ferramenta**: Utilização da **API Groq** para geração de textos criativos. Usar uma `GROQ_API_KEY` temporária ou que seja lida externamente.
*   **Prompt Engineering**: Para cada livro humano, gerar um **"gêmeo artificial"** usando um prompt que mistura estilos e temas de trechos diferentes (extraídas de partes distintas do mesmo livro original).
*   **Tratamento de Erros**: Implementar uma estratégia de para lidar com o Rate Limit (HTTP 429) do plano gratuito da API. Os prompts utilizados serão salvos em arquivos `.md` para reprodutibilidade.
*   **Fallback**: Em caso de falha irrecuperável na geração da API, a amostra humana correspondente será clonada para o corpus CA.

#### Fase 3: Modelagem Matemática (Regressão Linear)

Estimação do coeficiente **$\alpha$** de Zipf:

1.  **Contagem de Frequências**: Calcular as frequências das palavras  e considerar as  palavras mais frequentes.
2.  **Regressão Linear**: Realizar regressão linear pelo método dos mínimos quadrados no espaço **log-log** para a relação $\log(f)$ vs. $\log(r)$, onde $\log(f) = \log(C) - \alpha \cdot \log(r)$. O coeficiente $\alpha$ é retornado como o valor absoluto da inclinação.

#### Fase 4: Análise Estatística e Visualização

Comparação rigorosa entre CN e CA:

| Métrica | Descrição |
|---------|-----------|
| **Média e Desvio Padrão** | Calcular e comparar a média e o desvio padrão dos coeficientes $\alpha$ para ambos os corpora. |
| **Teste de Hipótese (Kolmogorov-Smirnov - KS)** | Aplicar o teste de `scipy.stats.ks_2samp` para verificar se as distribuições de $\alpha$ entre CN e CA são estatisticamente semelhantes (p-value < 0.05 indica distribuições distintas). |
| **Gráficos** | Gerar histogramas comparativos das distribuições de $\alpha$ (com médias e desvios padrão indicados) e gráficos de ajuste (linhas de tendência) no espaço log-log para um par de exemplo CN/CA, ambos salvos em arquivos PNG. |
| **Resultados** | Salvar os coeficientes $\alpha$ calculados, juntamente com os IDs do Gutenberg, em um arquivo CSV (`alphas.csv`) para análise posterior. |

````python
import os
import re
import time
import requests
from pathlib import Path
from groq import Groq
from dotenv import load_dotenv
from concurrent.futures import ThreadPoolExecutor
````

### Extração, processamento e limpeza dos livros Corpus Natural e obtenção do Corpus Artificial

````python
# Carrega a chave do Groq do arquivo .env
load_dotenv()

# Inicializa o cliente do Groq
try:
    client = Groq()
except Exception as e:
    print(f"Erro ao inicializar Groq. Verifique o arquivo .env: {e}")
    client = None

GUTERBERG_IDS = [
    # Clássicos absolutos e longos (Austen, Shelley, Stoker, Melville)
    84, 1342, 2701, 345, 1661, 98, 1400, 76, 120, 161,
    1260, 74, 148, 219, 514, 45, 215, 174, 768, 160,
    
    # Épicos e grandes narrativas (Dickens, Hugo, Dumas, Tolstói)
    4300, 2600, 135, 36, 1184, 5230, 158, 2814, 121, 105,
    110, 23, 3153, 328, 482, 153, 145, 550, 2828, 143,
    
    # Aventura, ficção científica clássica e realismo
    1259, 2604, 2610, 1399, 2800, 2554, 2197, 2638, 86, 119,
    1837, 3176, 1044, 3186, 1900, 13720, 34, 421, 1075, 118,
    
    # H.G. Wells, Júlio Verne, Jack London e outros romances
    83, 103, 164, 18857, 1011, 159, 35, 730, 917, 766,
    580, 1023, 786, 963, 821, 564, 1406, 28054, 134, 162,
    
    # Mais clássicos literários do Top 100-200 do Gutenberg
    344, 408, 423, 501, 521, 541, 551, 653, 700, 804,
    829, 863, 902, 968, 1009, 1056, 1146, 1155, 1200, 1300
]

def extrair_texto_principal(texto_bruto):
    """Remove a licença do Project Gutenberg (cabeçalho e rodapé)."""
    # Procura as marcações padrões que iniciam e terminam o livro real
    inicio_match = re.search(r'\*\*\* START OF TH(E|IS) PROJECT GUTENBERG.*?\*\*\*', texto_bruto)
    fim_match = re.search(r'\*\*\* END OF TH(E|IS) PROJECT GUTENBERG.*?\*\*\*', texto_bruto)
    
    if inicio_match and fim_match:
        # Retorna apenas o que está entre o cabeçalho e o rodapé
        return texto_bruto[inicio_match.end():fim_match.start()]
    return texto_bruto

def limpar_texto(texto):
    """Remove números, pontuações e deixa tudo em minúsculas (opcional, mas recomendado para Zipf)."""
    # Para Zipf, geralmente convertemos tudo para minúscula (lower) para não separar "The" de "the"
    texto = texto.lower()
    return re.sub(r'[^a-z\s]', '', texto)

def extrair_palavras(texto, limite=25000):
    """Garante que a obra tem o tamanho mínimo e retorna a string cortada."""
    palavras = texto.split()
    if len(palavras) < limite:
        return None
    return " ".join(palavras[:limite])

def obter_trechos_mistos(texto_25k):
    """Extrai 3 trechos de 400 palavras (início, meio e fim) do texto."""
    palavras = texto_25k.split()
    
    # Agora que removemos a licença, o início é realmente o início da história
    trecho_inicio = " ".join(palavras[0:400])
    trecho_meio = " ".join(palavras[12000:12400])
    trecho_fim = " ".join(palavras[24000:24400])
    
    return trecho_inicio, trecho_meio, trecho_fim

def gerar_gemeo_artificial(texto_25k, nome_livro):
    """Envia os trechos mistos ao Groq para gerar uma nova obra derivada."""
    if not client:
        return None
        
    t1, t2, t3 = obter_trechos_mistos(texto_25k)
    
    prompt = f"""
    You are a literary author specializing in pastiche, stylistics, and the fusion of themes.
    Below are three distinct excerpts taken from the work '{nome_livro}'.
    Your mission is to meticulously analyze the writing style, atmosphere, word choice, and underlying themes of each fragment.
    Then, create an "Artificial Twin" of that work: an original, complete, and compelling short story (2,000 words) in the language of the original work, organically blending and fusing the styles and themes observed in the three excerpts. The text must be fluid and purely literary.

    --- TRECHO 1 (Contexto Inicial) ---
    {t1}

    --- TRECHO 2 (Desenvolvimento) ---
    {t2}

    --- TRECHO 3 (Clímax/Conclusão) ---
    {t3}
    ---
    Gêmeo Artificial Gerado:
    """

    try:
        resposta = client.chat.completions.create(
            model="openai/gpt-oss-120b", 
            messages=[
                {"role": "system", "content": "Você é um escritor fantasma de alta literatura que cria textos perfeitamente mimetizados."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.85,
            max_tokens=3000 # Aumentado um pouco para garantir espaço para 2000 palavras
        )
        return resposta.choices[0].message.content
    except Exception as e:
        print(f"Erro na API Groq para o livro {nome_livro}: {e}")
        return None

def processar_livro(url):
    try:
        response = requests.get(url, timeout=10)
        if response.status_code != 200:
            return False

        id_livro = url.split('/')[-1].split('.')[0][2:] # Melhora a extração do ID: de "pg1000.txt" para "1000"
        
        texto_bruto = response.text
        # Passo NOVO e essencial: remover a licença
        texto_miolo = extrair_texto_principal(texto_bruto)
        texto_limpo = limpar_texto(texto_miolo)
        texto_25k = extrair_palavras(texto_limpo, limite=25000)

        if not texto_25k:
            print(f"Ignorado: {id_livro} não atingiu 25k palavras após limpeza.")
            return False

        Path("./CN").mkdir(parents=True, exist_ok=True)
        Path("./CA").mkdir(parents=True, exist_ok=True)

        path_original = f"CN/{id_livro}.txt"
        with open(path_original, mode="w", encoding="utf-8") as f:
            f.write(texto_25k)

        print(f"Gerando Gêmeo Artificial para {id_livro}...")
        texto_gemeo = gerar_gemeo_artificial(texto_25k, nome_livro=id_livro)

        if texto_gemeo:
            path_gemeo = f"CA/{id_livro}.txt"
            with open(path_gemeo, mode="w", encoding="utf-8") as f:
                f.write(texto_gemeo)
            print(f"-> Sucesso Absoluto: Original salvo em CN e Gêmeo salvo em CA ({id_livro}).")
            return True
        else:
            print(f"-> Sucesso parcial: Apenas o original de {id_livro} foi salvo em CN (Falha na IA).")
            return False

    except requests.exceptions.RequestException as e:
        print(f"Erro de conexão com {url}: {e}")
        return False

def iniciar_pipeline():
    template_url = "https://www.gutenberg.org/cache/epub/{ID}/pg{ID}.txt"
    urls = [template_url.format(ID=id) for id in GUTERBERG_IDS]

    obras_completas = 0

    with ThreadPoolExecutor(max_workers=2) as executor:
        resultados = executor.map(processar_livro, urls)
        for sucesso in resultados:
            if sucesso:
                obras_completas += 1
            # Pausa de segurança para a API da Groq não dar HTTP 429
            time.sleep(1) 

    print(f"\nPipeline Finalizado! Total de pares criados e salvos: {obras_completas}")


iniciar_pipeline()
````

## Limpeza do Corpus Artificial

````python
def limpar_corpus_artificial():
    """Lê os textos gerados pela IA, aplica a limpeza e sobrescreve os arquivos."""
    pasta_ca = Path("./CA")
    
    # Verifica se a pasta existe para evitar erros
    if not pasta_ca.exists():
        print("Pasta CA não encontrada. Certifique-se de que o pipeline gerou os arquivos.")
        return

    # Lista todos os arquivos .txt na pasta CA
    arquivos_ca = list(pasta_ca.glob("*.txt"))
    arquivos_processados = 0

    print(f"Iniciando a limpeza de {len(arquivos_ca)} arquivos no Corpus Artificial (CA)...")

    for caminho_arquivo in arquivos_ca:
        try:
            # 1. Lê o texto bruto gerado pela IA
            with open(caminho_arquivo, mode="r", encoding="utf-8") as f:
                texto_bruto = f.read()
            
            # 2. Aplica a função de limpeza existente (minúsculas, sem pontuação/números)
            texto_limpo = limpar_texto(texto_bruto)
            
            # 3. Sobrescreve o arquivo com o texto limpo
            with open(caminho_arquivo, mode="w", encoding="utf-8") as f:
                f.write(texto_limpo)
                
            arquivos_processados += 1
            
        except Exception as e:
            print(f"Erro ao limpar o arquivo {caminho_arquivo.name}: {e}")

    print(f"Limpeza Concluída! Total de livros artificiais limpos: {arquivos_processados}")

# Executa a limpeza
limpar_corpus_artificial()
````

````python
import pandas as pd
import collections
import numpy as np  
import matplotlib.pyplot as plt


pasta_ca = Path("./CA")

arquivos_ca = list(pasta_ca.glob("*.txt"))

for caminho_arquivo in arquivos_ca:
    try:
            # 1. Lê o texto bruto gerado pela IA
            with open(caminho_arquivo, mode="r", encoding="utf-8") as f:
                texto_bruto = f.read()
            
            words = texto_bruto.split()
            top = 100
            word_freq = collections.Counter(words)

            top_word_freq = word_freq.most_common(top)

            freq=pd.DataFrame(top_word_freq)

        
            x=np.log10(freq.index.values+1)

            y=np.log10(freq[1].values)
            cc=np.polyfit(x,y, deg=1);
            pol=np.poly1d(cc)
            plt.plot(x,y,'ro', x,pol(x), 'b')
            plt.xlabel('$s$');  plt.ylabel('$p(s)$');
            plt.show()
            print(pol)

            
    except Exception as e:
        print(f"Erro ao limpar o arquivo {caminho_arquivo.name}: {e}")

plt.show()
````
